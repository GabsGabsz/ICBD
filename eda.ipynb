Passos:
- Carregar dados
- Ver shape, colunas e tipos
- Contagem de valores nulos
- Estatísticas descritivas
- Distribuições e contagens para colunas-chave

```python
# 1) Importar bibliotecas e carregar o arquivo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

csv_path = r"d:\IC\banco_sbbrasil_2023_XLM_CSV\banco_sbbrasil_2023-20250326.csv"
# Ler com engine='python' e separador padrão ','
df = pd.read_csv(csv_path, sep=',', engine='python', encoding='utf-8')

print('Carregado em:', csv_path)
print('shape =', df.shape)
df.head()
```

```python
# 2) Colunas, tipos e nulos
print('Colunas ({}):'.format(len(df.columns)))
print(df.columns.tolist())

print('\nDtypes:')
print(df.dtypes)

print('\nNulos por coluna (count):')
nulls = df.isnull().sum()
print(nulls[nulls>0].sort_values(ascending=False).head(20))
```

```python
# 3) Estatísticas descritivas (numéricas e categóricas separadas)
num = df.select_dtypes(include=[np.number])
cat = df.select_dtypes(exclude=[np.number])

print('Numéricas: count, mean, std, min, 25%, 50%, 75%, max')
display(num.describe().T)

print('\nCategóricas (top 10 unique counts)')
display(cat.nunique().sort_values(ascending=False).head(20))
```

```python
# 4) Contagens e distribuições para colunas-chave (exemplos)
cols_cat = ['regiao','uf2','cidade','escolaridade']
for c in cols_cat:
    if c in df.columns:
        print('----', c, '----')
        display(df[c].value_counts().head(10))

# Distribuição de uma variável numérica exemplar
if 'idade' in df.columns:
    plt.figure(figsize=(8,4))
    df['idade'].dropna().astype(float).hist(bins=30)
    plt.title('Histograma - idade')
    plt.xlabel('idade')
    plt.show()

# Countplot de regiao (se existir)
if 'regiao' in df.columns:
    plt.figure(figsize=(6,3))
    sns.countplot(data=df, x='regiao')
    plt.title('Contagem por regiao')
    plt.show()
```
